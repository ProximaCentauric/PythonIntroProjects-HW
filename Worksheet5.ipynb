````python
# -------------------------------------
# Do not change the code below
# -------------------------------------

with open( "nc_county_data.pkle", 'rb') as handle:
    data = handle.read()
    
county_dict = pickle.loads(data)

# -------------------------------------
# Add your plotting solution below
# -------------------------------------
county_dict_short = {key: county_dict[key] for key in county_dict.keys()
       & {'Wake', 'Mecklenburg','Halifax'}}
Counties = list(county_dict_short.keys())
Counties.sort()
print(Counties)
USDollars = list(county_dict_short.values())
USDollars.sort()
print(USDollars)
colors = ["red","orange","green"]
explode = [0.05,0.05,0.05]
plt.figure(figsize=(15,5))
plt.suptitle("Median Income for Three NC Counties")
plt.subplot(121)
plt.barh(Counties,USDollars, color=colors)
plt.xlabel("US Dollars")
plt.subplot(122)
plt.pie(USDollars, labels=Counties, colors=colors, explode=explode, shadow=True, startangle = 90, autopct = '%1.1f%%',)
plt.show()
````

### Question 4: Visualizing Hippocampus Total Surface Area  (5 points)

<img src="ws5_q4_plot.png" width="350" style="float: right"/>

In the coding cell below, create histogram plot that visualizes the total surface area of the hippocampus subcortical brain region. You are provided a 2D array (or **matrix**) named **processed_hippo_data**, where 
- the first column is the surface area in millimeters squared ($mm^2$) of the **left** hippocampus, and 
- the second column is the surface area of the **right** hippocampus also in millimeters squared ($mm^2$). 

To recieve full credit your plotting solution must use the <a href="https://matplotlib.org/">Matplotlib</a> Python library and create a plotting solution that is **visually identical** to the `plot shown on the right`. You may assume:
- The colors used to generate the histogram plots are blue and red.
- The plot figure size is 8 inches in width and 5 inches in height.
- The opacity (alpha value) for both histogram plots is 0.5
- The total number of equal size bins in each histogram plot is 150.

Hints:
- Matplotlib <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html">hist</a> 



````python
# -------------------------------------
# Do not change the code below
# -------------------------------------

hippo_data = np.genfromtxt('hippo.csv',delimiter=',')

# create an imputation model and replace missing values
model = impute.SimpleImputer(missing_values=np.nan, strategy='mean')
processed_hippo_data = model.fit_transform( hippo_data ).astype( int )

# -------------------------------------
# Add your plotting solution below
# -------------------------------------
left = processed_hippo_data[:,0]
right = processed_hippo_data[:,1]
plt.figure(figsize=(8,5))
plt.title("Hippocampus Analysis for 2,776 Subjects")
plt.hist(left, bins=150, facecolor="b", alpha=0.5)
plt.hist(right, bins=150, facecolor="r", alpha=0.5)
plt.xlabel("Total Surface Area ($mm^2$)")
plt.ylabel("Total Number of Occurances")
plt.legend([
    "left hippocampus",
    "right hippocampus"
])
plt.show()
````
