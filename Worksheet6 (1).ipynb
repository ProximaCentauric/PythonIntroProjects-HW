### Question 3: Boston Red Sox (5 points)

<img src="ws6_q3_plot.png" width="475" style="float: right"/>

In the coding cell below, create a combined line and pie plot that visualizes the ``mean`` number of **double**, **triple**, and **homerun** hits by the **Boston Red Sox** between 2000 and 2010. 

To recieve full credit your plotting solution must use:
1. the data provided in the **team_df** Panda dataframe,
2. the <a href="https://matplotlib.org/">Matplotlib</a> Python library (i.e., **do not** use the Panda plotting capabilities), and 
3. a plotting solution that is **visually identical** to the `plot shown on the right`. 

You may assume:
- The colors used to generate plots are orange (doubles), red (triples), and green (homerunes)
- The plot figure size is 15 inches in width and 5 inches in height.
- The explode values for the pie plot are all 0.05
- The pie plot does use shadding and the start angle is 90.
- The pie plot auto percent format uses one decimal place.

Dataframe Hints:
- <a href="https://pandas.pydata.org/docs/user_guide/indexing.html">Selecting, Slicing, and Extracting</a> 
- Convert to <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_numpy.html">Numpy Array</a>
- Convert to <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.tolist.html">Python List</a>
- Data selection using <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html">between</a>


````python
# -------------------------------------
# Add your plotting solution below
# -------------------------------------

BRSdf = df.loc[df["team name"] == "Boston Red Sox"]
RSdf = BRSdf.loc[BRSdf["year"].between(2000,2010)]
year = RSdf.loc[:,"year"]
doubles = RSdf.loc[:,"doubles"]
triples = RSdf.loc[:,"triples"]
homeruns = RSdf.loc[:,"homeruns"]
colors = ["orange","red","green"]
explode = [0.05,0.05,0.05]
basenum = [doubles.mean(),triples.mean(),homeruns.mean()]
namenum = ["doubles", "triples", "homeruns"]


plt.suptitle("Boston Red Sox Hitting Data (2000 to 2010)")
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(year, doubles, "o-", color="orange")
plt.plot(year, triples, "ro-")
plt.plot(year, homeruns, "go-")
plt.xticks(year)
plt.legend(["doubles","triples","homeruns"])
plt.subplot(122)
plt.pie(basenum, labels=namenum,colors=colors, explode=explode, shadow=True, startangle=90,autopct = '%1.1f%%')
plt.show()
````

### Question 4:  The Steroid and Post-Steriod Era Comparison (5 Points)

<img src="ws6_q4_plot.png" width="475" style="float: right"/>

In the coding cell below, create a combined histogram plot that visualizes the ``median`` number of **homeruns** for each team in the steriod era (between 1994 and 2005) and in the post-steriod era (between 2006 and 2015). 

To recieve full credit your plotting solution must use:
1. the data provided in the **team_df** Panda dataframe,
2. the <a href="https://matplotlib.org/">Matplotlib</a> Python library (i.e., **do not** use the Panda plotting capabilities), and 
3. a plotting solution that is **visually identical** to the `plot shown on the right`. 

You may assume:
- The plot figure size is 10 inches in width and 20 inches in height.
- Steriod era data is grouped by `team name` and `homeruns` then aggregated by `median` operation.
- Likewise, post-steriod era data is also grouped by `team name` and `homeruns` then aggregated by `median` operation.

Dataframe Hints:
- <a href="https://pandas.pydata.org/docs/user_guide/indexing.html">Selecting, Slicing, and Extracting</a> 
- Convert to <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_numpy.html">Numpy Array</a>
- Convert to <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.tolist.html">Python List</a>
- Data selection using <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html">between</a>
- Data aggregating using <a href="https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.aggregate.html">groupby and aggregate</a>


````python
# -------------------------------------
# Add your plotting solution below
# -------------------------------------

Sdf = df[df["year"].between(1994,2005)]
Pdf = df[df["year"].between(2006,2014)]

dd1 = Sdf.groupby(['team name','homeruns']).agg("median").reset_index()
dd2 = Pdf.groupby(['team name','homeruns']).agg('median').reset_index()

plt.figure(figsize=(10,20))
plt.subplot(211)
plt.title("Steroid Era (1994 to 2005)")
plt.barh("team name", "homeruns", data=dd1)
plt.xlabel("Median # of Homeruns")
plt.subplot(212)
plt.title("Post-Steroid Era (2006 to 2014)")
plt.barh("team name", "homeruns", data=dd2)
plt.xlabel("Median # of Homeruns")
plt.show()
````
